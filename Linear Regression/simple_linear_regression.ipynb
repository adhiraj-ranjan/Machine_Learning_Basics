```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import random

n_1 = np.array(random.sample([i for i in range(1000)], 10))
n_2 = (np.array(n_1) *2)/5
```

```python
# n_2 = n_2.astype("int64")
```

```python
n_2
```

```python
dataset = pd.DataFrame({"n_1": n_1, "n_2": n_2})
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(x = "n_1", y = "n_2", data = dataset)
plt.show()
```

```python
 x = dataset[["n_1"]]
 y = dataset["n_2"]
```

```python
x.ndim, y.ndim
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.15, random_state=42)
```

```python
train_x.shape, train_y.shape, test_x.shape, test_y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(train_x, train_y)
```

```python
accuracy = lr.score(test_x, test_y)*100
accuracy
```
